```python
from os import path
from wordcloud import WordCloud

#d = path.dirname(__file__)

# Read the whole text.
document_text = open('test.txt', 'r')
text_string = document_text.read().lower()
#text = open(path.join(d, 'test.txt')).read()
type(text_string)
```

```python
#cleanx= text_string.strip('http\S+|www.\S+', '', case=False)
import re
text = re.sub(r'http\S+|www.\S+', '', text_string, flags=re.MULTILINE)
```

```python
import csv

with open("output.csv",'w') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerow(text)
```

```python
#data['Clean_Tweets'] = data['bodyText'].str.replace('http\S+|www.\S+', '', case=False)

# Generate a word cloud image
wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
