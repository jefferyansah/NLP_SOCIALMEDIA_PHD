```python
# Excel column name should be "column_a" and "column_b"
import pandas as pd
import numpy as np
df = pd.read_excel('output.xlsx', sheetname='Sheet1')
df
```

```python
accuracy_list = []
for index, row in df.iterrows():
    text_list = []
    text_split = row['column_a'].split(' ') if not pd.isnull(row['column_a']) else []
    print(text_split)
    for text in text_split:
        counter = row['column_b'].count(text)
        if counter:
            text_list.append(text)
    accuracy_list.append("{}%".format((len(text_list) * 100)/ len(text_split)))
   # accuracy_list.append("{}%".format((len(text_split) * 100)/ len(text_list)))
df['accuracy'] = accuracy_list
df
```
