```python
import matplotlib.pyplot as plt
import pylab



x =[1,2,3,4,5]
svm =[86.27,75.55,77.47,82.69, 87.72]
vbm =[71.27,71.16,73.33,69.11,74.86]
emvbm =[65.03,65.89,66.06,70.48,67.65]
my_xticks =[ 'Ω1', 'Ω2', 'Ω3', 'Ω4', 'Ω5']
plt.xticks(x, my_xticks,)

fig = plt.figure(1, figsize=(3, 5))

plt.ylim(ymax = 90, ymin = 60) #set limist for axes
pylab.plot(x, svm, '-g',  label='Our Model', marker='x', linestyle='-')
pylab.plot(x, vbm, '-b',  label='Comparism Model', marker='^', linestyle='-')
pylab.plot(x, emvbm, '-r',  label='baseline', marker='^', linestyle='-')
plt.xticks(x, my_xticks,)

plt.xlabel("Prediction Months")
plt.ylabel("Model Accuracy")
plt.title(' (a) Average Acurracy: LeadTime 1')
plt.grid()
#pylab.plot(days, coalmine, '-k', label='#coal')
#pylab.plot(days, keepit, '-g', label='#coalpower', marker='.')

pylab.legend(loc='upper center', ncol=2, bbox_to_anchor=(0.5, 0.99), fontsize='small')
#pylab.plot(my_xticks, fontsize='small')
plt.show()
fig.savefig('d1.png', bbox_inches='tight')
```
