```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np


plt.figure(figsize=(8,10))
ax = plt.gca()
im = ax.imshow(harvest)

ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))

ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)
ax.set_ylabel('Tree-ID', labelpad=-5, fontsize=12)
ax.set_xlabel('Time Windows', labelpad=-2)
#fontweight='bold'
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
ax.tick_params(axis='x', which='major', pad=-4)
ax.tick_params(axis='y', which='major', pad=1)

ax.set_title("Tree Data Grid Heatmap of Bursty Windows(T=60min)", fontsize=12)
fig.tight_layout()

# create an axes on the right side of ax. The width of cax will be 5%
# of ax and the padding between cax and ax will be fixed at 0.05 inch.
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%", pad=0.12)

#fig, ax = plt.subplots(figsize=(8,6))
plt.colorbar(im, cax=cax)
plt.savefig('at1.png',bbox_inches='tight', dpi = 1000)
```
