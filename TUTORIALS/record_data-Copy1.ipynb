```python
import matplotlib.pyplot as plt
import pylab



x =[1,2,3,4,5]
knn =[84.27,79.55,80.34,82.69, 87.71]
svm =[86.27,78.55,79.47,82.69, 87.72]
bsl =[78.96,74.55,74.41,76.22, 78.86]
vbm =[71.27,71.16,73.33,69.11,74.86]
cgraph =[69.78,76.89,78.23,78.22,79.18]
emvbm =[65.03,65.89,66.06,70.48,67.65]
my_xticks =[ '1', '2', '3', '4', '5']
plt.xticks(x, my_xticks,)

fig = plt.figure(2, figsize=(8, 8))

plt.ylim(ymax = 90, ymin = 60) #set limist for axes
pylab.plot(x, knn, '-g',  label='XCPT1', marker='o', linestyle='-')
pylab.plot(x, svm, '-m',  label='XCPT2', marker='x', linestyle='-')
pylab.plot(x, bsl, '-r',  label='GSR Baseline', marker='H', linestyle='-')
pylab.plot(x, vbm, '-c',  label='VBM_emb', marker='d', linestyle='-')
pylab.plot(x, emvbm, '-k',  label='VMB_kw', marker='^', linestyle='-')
pylab.plot(x, cgraph, '-b',  label='Cgraph', marker='s', linestyle='-')
plt.xticks(x, my_xticks ,fontsize=12)

plt.xlabel("Test Windows ",fontsize=16 )
plt.ylabel("Model Accuracy" ,fontsize=16)
plt.title(' (d) Average Acurracy: LeadTime 1' ,fontsize=17)
#plt.grid()


pylab.legend(loc='upper center', ncol=3, bbox_to_anchor=(0.45, 0.99), fontsize=12)
#pylab.plot(my_xticks, fontsize='small')
plt.show()

fig.savefig('acc1.png', bbox_inches='tight')
```
